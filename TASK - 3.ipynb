```python
#TASK - 3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Loading dataset...")
df = pd.read_csv("cleaned_products.csv")

print("Creating box plot for price distribution...")
plt.figure(figsize=(8, 6))
sns.boxplot(y="Price", data=df, color="lightblue")
plt.title("Price Distribution of Books")
plt.ylabel("Price")
plt.grid(True, axis="y")
plt.savefig("price_boxplot_.png")
plt.show()
plt.close()

print("Creating histogram for price distribution...")
plt.figure(figsize=(8, 6))
sns.histplot(df["Price"], bins=10, kde=True, color="skyblue")
plt.title("Price Distribution of Books")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(True, axis="y")
plt.savefig("price_histogram_.png")
plt.show()
plt.close()

print("Creating bar chart for average price by rating...")
avg_price_by_rating = df.groupby("Rating")["Price"].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x="Rating", y="Price", data=avg_price_by_rating, color="salmon")
plt.title("Average Price by Rating")
plt.xlabel("Rating")
plt.ylabel("Average Price")
plt.grid(True, axis="y")
plt.savefig("avg_price_by_rating_.png")
plt.show()
plt.close()

print("Creating scatter plot for price vs rating...")
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Rating", y="Price", data=df, color="green", alpha=0.6)
plt.title("Price vs Rating of Books")
plt.xlabel("Rating")
plt.ylabel("Price")
plt.grid(True)
plt.savefig("price_vs_rating_.png")
plt.show()
plt.close()

print("Key Insights:")
print("- The box plot shows the price distribution, with potential outliers indicating premium-priced books.")
print("- The histogram reveals a skewed price distribution, with most books priced in a lower range.")
print(f"- Average prices by rating suggest {'minimal' if df.groupby('Rating')['Price'].mean().std() < 5 else 'notable'} variation.")
print("- The scatter plot indicates no strong visual correlation between price and rating.")
```
